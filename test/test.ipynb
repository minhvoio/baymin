```python
import pdfplumber

file_path = "./data/A1 _ 0103.pdf"

all_text = ""
with pdfplumber.open(file_path) as pdf:
    for page in pdf.pages:
        all_text += page.extract_text() + "\n"

print(all_text[:2000])
```

```python
from ollama_ocr import OCRProcessor
file_path = "./data/A1 _ 0103_1-3.pdf"
# Initialize OCR processor
ocr = OCRProcessor(model_name='llama3.2-vision:11b', base_url="http://localhost:11434/api/generate")  # You can use any vision model available on Ollama
# you can pass your custom ollama api

# Process an image
result = ocr.process_image(
    image_path=file_path, # path to your pdf files "path/to/your/file.pdf"
    format_type="markdown",  # Options: markdown, text, json, structured, key_value
    custom_prompt="Extract all text, focusing on dates and names.", # Optional custom prompt
    language="English" # Specify the language of the text (New! üÜï)
)
print(result)
```
